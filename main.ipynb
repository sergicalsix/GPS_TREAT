```python
import pandas as pd
import numpy as np
import folium

df = pd.read_csv("/Users/shibuya/Downloads/before_gps.csv")
df_before = pd.read_csv("/Users/shibuya/Downloads/before_gps (1).csv")
df_after = pd.read_csv("/Users/shibuya/Downloads/after_gps (1).csv")

df_before.columns = ['lat','long']
df_after.columns = ['lat','long']

df = df.to_numpy()
before_array = df_before.to_numpy()
#before_array[:,0] += np.random.randn(len(before_array))  * 0.01
#before_array[:,1] += np.random.randn(len(before_array))  * 0.01
after_array = df_after.to_numpy()
```

```python
folium_map = folium.Map(location=[before_array[0][0], before_array[0][1]], zoom_start=15)

for i, location in enumerate(before_array):
    if i % 100 == 0:    
        folium.Marker(
            location=[location[0], location[1]],
            icon=folium.Icon(color='red')
        ).add_to(folium_map)

    
folium_map 
```

```python
folium_map = folium.Map(location=[after_array[0][0], before_array[0][1]], zoom_start=15)

for i, location in enumerate(after_array):
    if i % 100 == 0:    
        folium.Marker(
            location=[location[0], location[1]],
            icon=folium.Icon(color='blue')
        ).add_to(folium_map)
    
folium_map 
```
